```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis = 1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr_y_train_pred = lr.predict(x_train)
lr_y_test_pred = lr.predict(x_test)
```

```python
lr_y_train_pred
```

```python
lr_y_test_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train, lr_y_train_pred)
lr_train_r2 = r2_score(y_train, lr_y_train_pred)

lr_test_mse = mean_squared_error(y_test, lr_y_test_pred)
lr_test_r2 = r2_score(y_test, lr_y_test_pred)
```

```python
lr_results = pd.DataFrame (['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=0)
rf.fit(x_train,y_train)
rf_y_train_pred = rf.predict(x_train)
rf_y_test_pred = rf.predict(x_test)
```

```python
rf_y_train_pred
```

```python
rf_y_test_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train, rf_y_train_pred)
rf_train_r2 = r2_score(y_train, rf_y_train_pred)

rf_test_mse = mean_squared_error(y_test, rf_y_test_pred)
rf_test_r2 = r2_score(y_test, rf_y_test_pred)
```

```python
rf_results = pd.DataFrame (['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
rf_results
```

```python
from sklearn.linear_model import Ridge
r = Ridge()
r.fit(x_train,y_train)
```

```python
r_y_train_pred = r.predict(x_train)
r_y_test_pred = r.predict(x_test)
```

```python
r_y_train_pred
```

```python
r_y_test_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
r_train_mse = mean_squared_error(y_train, r_y_train_pred)
r_test_mse = mean_squared_error(y_test, r_y_test_pred)

r_train_r2 = r2_score(y_train, r_y_train_pred)
r_test_r2 = r2_score(y_test, r_y_test_pred)
```

```python
r_results = pd.DataFrame(['Ridge Regression', r_train_mse, r_train_r2, r_test_mse, r_test_r2]).transpose()
r_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
r_results
```

```python
from sklearn.linear_model import Lasso
l = Lasso()
l.fit(x_train, y_train)
l_y_train_pred = l.predict(x_train)
l_y_test_pred = l.predict(x_test)
```

```python
l_y_train_pred 
```

```python
l_y_test_pred
```

```python
l_results = pd.DataFrame(['Lasso Regression', l_train_mse, l_train_r2, l_test_mse, l_test_r2]).transpose()
l_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
l_results
```
